```python
test_index = '801'
trades = trade_results[test_index]
print(trades)

hedge_ratio = trade_results[test_index+'_beta'][0]
print(f"\nhedge_ratio: {hedge_ratio}")
```

```python
def cal_spread( x, y,beta, is_norm=False):
    resid = y - x * beta
    resid = (resid - resid.mean()) / resid.std() if is_norm is True else resid
    return resid
```

```python
for i in trades.index:    
    trade_time = trades.loc[i,'trade_time']
    close_time = trades.loc[i,'close_time']
    long_short = trades.loc[i,'long_short']
    
    if long_short==1:
        trade_price =  y.loc[y.date==trade_time,'close'] - x.loc[x.date==trade_time,'close'] * hedge_ratio
        print(trade_price)
                                       
        y_profit = y.loc[y.date==close_time,'close'].values[0] -  y.loc[y.date==trade_time,'close'].values[0]
        x_profit = (x.loc[x.date==trade_time,'close'].values[0] - x.loc[x.date==close_time,'close'].values[0]) * hedge_ratio
    else:
        y_profit = y.loc[y.date==trade_time,'close'].values[0] -  y.loc[y.date==close_time,'close'].values[0]
        x_profit = (x.loc[x.date==close_time,'close'].values[0] - x.loc[x.date==trade_time,'close'].values[0]) * hedge_ratio
        
    profit = y_profit + x_profit
    print(f"profit: {i}: {profit}")
```
